```python
def shuffle(df):
    col = df.columns
    val = df.values
    shape = val.shape
    val_flat = val.flatten()
    np.random.shuffle(val_flat)
    return pd.DataFrame(val_flat.reshape(shape),columns=col)
```

# Logistic Regression

```python
#us_data_dest.columns
us_lr = us_data#[us_data_dest.columns]
us_lr["booking_date"] = us_lr["date_time"].str.extract('(....-..-..)', expand = True)
us_lr["formatted_booking_date"] = pd.to_datetime(us_lr.booking_date, format="%Y/%m/%d")
us_lr["formatted_ci_date"] = pd.to_datetime(us_lr.srch_ci, format="%Y/%m/%d",errors="coerce")
us_lr["formatted_co_date"] = pd.to_datetime(us_lr.srch_co, format="%Y/%m/%d",errors="coerce")
us_lr["num_days"] = (us_lr.formatted_ci_date - us_lr.formatted_booking_date).dt.days
```

```python
def logistic_regression(df_lr, col_list,sample_size):
    for c in col_list:
        df_lr[c] = df_lr[c].astype('category').cat.codes    
    
    X = np.array(df_lr[col_list])
    y = np.array(df_lr["is_booking"])

    #X_sample = X[idx]
    #y_sample = y[idx]
    
    pos= y == 1
    neg = y == 0

    X_pos_0 = X[pos,0]
    X_pos_1 = X[pos,1]
    X_neg_0 = X[neg,0]
    X_neg_1 = X[neg,1]
    
    N1 = X_pos_0.size
    N2 = X_neg_0.size

    random.seed()
    idx1=random.sample(range(N1),sample_size)
    idx2=random.sample(range(N2),sample_size)

    fig, ax = plt.subplots(figsize=(12, 9))
    scatter(X_neg_0[idx2], X_neg_1[idx2], marker='x', c='lightcoral',s=10)
    scatter(X_pos_0[idx1], X_pos_1[idx1], marker='o', c='darkslategrey',s=10)
    xlabel(col_list[0])
    ylabel(col_list[1])
    legend(['Book', 'Click'])
    show()
```

```python
def multiple_plots(df_lr,sample_size, full_list,col1,col2,metric):

    dim = len(full_list)
    fig, ax = plt.subplots(figsize=(dim*2, dim*1.5))

    for i in range(0,dim):
        for j in range(0,dim):
            if i!=j:
                col_list = [full_list[i],full_list[j]]
                for c in col_list:
                    df_lr[c] = df_lr[c].astype('category').cat.codes
                X = np.array(df_lr[col_list])
                y = np.array(df_lr[metric])

                #X_sample = X[idx]
                #y_sample = y[idx]

                pos= y == 1
                neg = y == 0

                X_pos_0 = X[pos,0]
                X_pos_1 = X[pos,1]
                X_neg_0 = X[neg,0]
                X_neg_1 = X[neg,1]

                N1 = X_pos_0.size
                N2 = X_neg_0.size

                random.seed(1)
                idx1=random.sample(range(N1),sample_size)
                idx2=random.sample(range(N2),sample_size)

                plt.subplot(dim, dim, j*dim+i+1)


                
                scatter(X_pos_0[idx1], X_pos_1[idx1], marker='o', c=col1,s=15)
                scatter(X_neg_0[idx2], X_neg_1[idx2], marker='x', c=col2,s=15,alpha=0.5)
                #xlabel(col_list[0])
                #ylabel(col_list[1])
    #legend(['Book', 'Click'],loc="upper center")

    show()
```
